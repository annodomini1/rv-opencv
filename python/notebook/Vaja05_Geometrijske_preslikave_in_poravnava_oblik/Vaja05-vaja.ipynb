# Vaja 5: Geometrijske preslikave in poravnava oblik

## Navodila
S prostorskimi ali geometrijskimi preslikavami $\mathcal{T}: \mathbb{R}^2 \rightarrow \mathbb{R}^2$ oz. $\mathcal{T}: \mathbb{R}^3 \rightarrow \mathbb{R}^3$ preslikamo slikovne elemente 2D slik $(x,y)$ oz. 3D slik $(x,y,z)$ na nove lokacije $(u,v)$ oz. $(u,v,w)$, njihove sivinske vrednosti pa pri tem ohranimo. Na ta način lahko izvedemo povečavo oz. pomanjšavo (skaliranje), premik (translacijo), zasuk (rotacijo), pa tudi številne druge linearne in nelinearne geometrijske preslikave slik. Poljubno geometrijsko preslikavo za 2D oz. 3D sliko zapišemo kot:

$$(u,v) = \mathcal{T}(x,y)\quad \mbox{oz.}\quad (u,v,w) = \mathcal{T}(x,y,z)\,.$$

Najbolj splošna linearna geometrijska preslikava je **afina preslikava**, ki omogoča poljubno skaliranje, translacijo, rotacijo in strig. Afina preslikava je v 2D določena s 6, v 3D pa z 12 parametri:

$$
\left[ 
	\begin{array}{c} u \\ v \\ 1 \end{array} 
	\right] = \underbrace{\left[ 
	\begin{array}{ccc} a_{11} & a_{12} & t_x \\ a_{21} & a_{22} & t_y \\ 0 & 0 & 1 \end{array}
	\right]}_{\mathcal{T}_{afina}}\,
	\left[ 
	\begin{array}{c} x \\ y \\ 1 \end{array}
	\right]
	\quad \mbox{oz.}\quad 
	\left[ 
	\begin{array}{c} u \\ v \\ w \\ 1 \end{array} 
	\right] = \underbrace{\left[ 
	\begin{array}{cccc} a_{11} & a_{12} & a_{13} & t_x \\ a_{21} & a_{22} & a_{23} & t_y \\
	a_{31} & a_{32} & a_{33} & t_z \\ 0 & 0 & 0 & 1 \end{array}
	\right]}_{\mathcal{T}_{afina}}\,
	\left[ 
	\begin{array}{c} x \\ y \\ z \\ 1 \end{array}
	\right]
$$

kjer parametri $t_x, t_y$ in $t_z$ določajo translacijo v $x,y$ in $z$ smeri slike, parametri $a_{ij}$ pa skaliranje, rotacijo in strig. Matriko za afino preslikavo lahko sestavimo z zaporednim množenjem homogenih matrik posameznih elementarnih preslikav v želenem vrstnem redu, naprimer:

$$
\mathcal{T}_{afina} = \mathcal{T}_{trans}\,\mathcal{T}_{strig}\,\mathcal{T}_{rot}\,\mathcal{T}_{skal}\,.
$$

V praksi se pogosto uporabljata **toga preslikava**, ki jo dobimo z združevanjem rotacijske in translacijske elementarne preslikave in **podobnostna preslikava**, ki jo dobimo z združevanjem toge preslikave in skaliranja.

### Geometrijska poravnava
Geometrijska poravnava dveh oblik je proces iskanja optimalne geometrijske preslikave $\mathcal{T}$, ki obliki preslika tako, da se iste strukture oz. značilnice oblike nahajajo v enakem ali čim bolj »podobnem« položaju. Značilnice 2D oz. 3D oblik so pogosto kar oblaki točk, zato se za mero podobnosti med dvema oblikama pogosto uporablja povprečna kvadratna Evklidska razdalja med pripadajočimi oz. korespondenčnimi točkami. Na osnovi množice $K$ parov korespondenčnih točk lahko določimo preslikavo $\mathcal{T}$ tako, da se preslikane točke referenčne oblike $\mathcal{T}(x_i,y_i)$ čim bolje prilegajo točkam vhodne oblike $(u_i,v_i)$ in obratno:

$$
(u_i,\,v_i) \leftrightarrow \mathcal{T}(x_i,\,y_i) \quad\mbox{oz.}\quad  (x_i,\,y_i) \leftrightarrow \mathcal{T}^{-1}(u_i,\,v_i)\,.
$$

Glede na število parov korespondenčnih točk in vrsto preslikave lahko točke preslikamo tako, da se popolnoma prekrivajo, tj. $\mathcal{T}(x_i,\,y_i)=(u_i,\,v_i)$, kar imenujemo **interpolacijska poravnava**, ali pa tako, da se točke le približno prekrijejo, tj. $\mathcal{T}(x_i,\,y_i) \approx (u_i,\,v_i)$, kar imenujemo **aproksimacijska poravnava**.

**Afino interpolacijsko poravnavo** med 2D oblikami lahko enolično določimo, če poznamo natanko tri pare ($K=3$) nekolinearnih točk $(u_i,\,v_i) \leftrightarrow (x_i,y_i)$, poravnavo med 3D oblikami pa če poznamo natanko šest parov ($K=6$) nekolinearnih točk. V 2D afino interpolacijsko poravnavo $\mathcal{T}$ med tremi pari točk zapišemo kot:

$$
\mathcal{T} = \left[ 
	\begin{array}{ccc} a_{11} & a_{12} & t_x \\ a_{21} & a_{22} & t_y \\ 0 & 0 & 1 \end{array}
	\right]\, = \,
	\left[ 
	\begin{array}{ccc} u_1 & u_2 & u_3 \\ v_1 & v_2 & v_3 \\ 1 & 1 & 1 \end{array}
	\right] \,
	\left[ 
	\begin{array}{ccc} x_1 & x_2 & x_3 \\ y_1 & y_2 & y_3 \\ 1 & 1 & 1 \end{array}
	\right]^{-1}\,.
$$

**Afino aproksimacijsko poravnavo** lahko med 2D oblikami izvedemo, kadar poznamo več kot tri pare ($K>3$), med 3D oblikami pa več kot šest parov ($K>6$) korespondenčnih točk. V tem primeru dobimo predoločen sistem enačb, zato korespondenčne točke lahko poravnamo le približno (aproksimacijsko). To storimo tako, da minimiziramo povprečno kvadratno Evklidsko razdaljo med korespondenčnimi točkami: 

$$
	\Sigma^2 = \frac{1}{K} \sum_{i=1}^{K} \| (u_i,\,v_i) - \mathcal{T}(x_i,\,y_i) \|^2 \,.
$$

Optimalne vrednosti neznanih parametrov preslikave dobimo tako, da odvode povprečne kvadratne Evklidske razdalje $\Sigma^2$ po vseh parametrih postavimo na nič in dobimo sistem enačb za reševanje. Za 2D afino aproksimacijsko poravnavo dobimo naslednji sistem enačb:

$$
\underbrace{\left[ 
	\begin{array}{cccccc} 
	\overline{xx} & \overline{xy} & \overline{x} & 0 & 0 & 0 \\ 
	\overline{xy} & \overline{yy} & \overline{y} & 0 & 0 & 0 \\ 
	\overline{x} & \overline{y} & 1 & 0 & 0 & 0 \\
	0 & 0 & 0 & \overline{xx} & \overline{xy} & \overline{x} \\ 
	0 & 0 & 0 & \overline{xy} & \overline{yy} & \overline{y} \\ 
	0 & 0 & 0 & \overline{x} & \overline{y} & 1 \\
	\end{array}
	\right]}_{\mathcal{P}_{xy}}\,
	\underbrace{\left[ 
	\begin{array}{c} 
	a_{11} \\ a_{12} \\ t_x \\ a_{21} \\ a_{22} \\ t_y \end{array}
	\right]}_{\mathbf{t}} \, = \,
	\underbrace{\left[ 
	\begin{array}{c} 
	\overline{ux} \\ \overline{uy} \\ \overline{u} \\ 
	\overline{vx} \\ \overline{vy} \\ \overline{v} \end{array}
	\right]}_{\mathbf{p}_{uv}}\,,
$$

kjer elementi s črto označujejo povprečne vrednosti $\overline{vx} = \frac{1}{K} \sum_{i=1}^K u_i x_i$ oz. $\overline{x} = \frac{1}{K} \sum_{i=1}^K x_i$. Zgornji sistem lahko zapišemo v matrični obliki kot $\mathcal{P}_{xy}\,\mathbf{t} = \mathbf{p}_{uv}$, kjer vektor $\mathbf{t}$ predstavlja vektorski zapis parametrov afine preslikave $\mathcal{T}$. Parametre afine preslikave dobimo z reševanjem sistema $\mathbf{t} = \mathcal{P}_{xy}^{-1}\,\mathbf{p}_{uv}$.

Pri poravnavi oblik korespondenčne točke običajno niso vnaprej znane in jih je potrebno določiti v procesu poravnave oblik. Ena od uveljavljenih metod za določanje korespondenčnih točk je **metoda iterativno najbližje točke** (ang.: iterative closest point (ICP)), pri kateri v vsakem koraku $k$ določimo korespondenčne točke z iskanjem najbližjih parov točk $(u_i,\,v_i) \leftrightarrow (x_i,\,y_i)$ in nato izračunamo interpolacijsko ali aproksimacijsko poravnavo $\mathcal{T}_k$ med temi pari točk. Korespondenčne pare točk določimo tako, da vsaka točka $(u_i,\,v_i)$ pripada kvečjemu enemu paru $(x_i,\,y_i)$. Premično obliko preslikamo s $\mathcal{T}_k$ in ponovimo korak. Postopek ponavljamo dokler je relativna sprememba $\|\mathcal{T}_k - I\|_{\infty} > \varepsilon$ oz. dokler ne izvedemo maksimalno število korakov $k_{max}$. Če so preslikave $\mathcal{T}_k$ linearne, potem lahko preslikavo med oblikama po $k$-tem koraku določimo z zaporednim matričnim množenjem vmesnih preslikav:

$$
\mathcal{T} = \mathcal{T}_k\,\mathcal{T}_{k-1}\,\mathcal{T}_{k-2}\,\cdots\,\mathcal{T}_{1}\,.	
$$

Funkcija vrne homogeno matriko `oMat2D`, ki ima dimenzije $3 \times 3$. Preverite delovanje funkcije tako, da točke izbrane oblike preslikate s poljubnimi parametri striga, rotacije, translacije in skaliranja, nato pa še s poljubno 2D afino preslikavo. Pri tem poskrbite, da bodo točke zapisane v homogenih koordinatah. Vhodne in preslikane točke oblike prikažite z ukazom `matplotlib.pylab.plot()`.

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
def transAffine2D(iScale=(1, 1), iTrans=(0, 0), iRot=0, iShear=(0, 0)):
    """Funkcija za poljubno 2D afino preslikavo"""
    iRot = iRot*np.pi/180
    oMatScale = np.array(((iScale[0],0,0),(0,iScale[1],0),(0,0,1)))
    oMatTrans = np.array(((1,0,iTrans[0]),(0,1,iTrans[1]),(0,0,1)))
    oMatRot = np.array(((np.cos(iRot),-np.sin(iRot),0),(np.sin(iRot),np.cos(iRot),0),(0,0,1)))
    oMatShear = np.array(((1,iShear[0],0),(iShear[1],1,0),(0,0,1)))
    oMat2D = np.dot(oMatTrans, np.dot(oMatShear, np.dot(oMatRot, oMatScale)))
    
    return oMat2D
````

````python
#ce je matrika dimenzija < (3x3) funkcija doda trji solpec enic
def addHomCoord2D(iPts):
    if iPts.shape[-1] == 3:
        return iPts
    iPts = np.hstack((iPts, np.ones((iPts.shape[0], 1))))
    return iPts
````

Obe matriki sta dimenzij $K \times 2$, kjer je $K \geq 3$. Funkcija vrne 2D afino preslikavo v obliki homogene matrike `oMat2D`, ki ima dimenzije $3 \times 3$. Preverite delovanje funkcije tako, da **vse točke** izbrane oblike preslikate s poljubnimi parametri 2D afine preslikave in preverite ali funkcija vrne preslikavo v `oMat2D`, ki je enaka vaši 2D afini preslikavi.

````python
def mapAffineApprox2D(iPtsRef, iPtsMov):
    """Afina aproksimacijska poravnava"""
    iPtsRef = np.matrix(iPtsRef)
    iPtsMov = np.matrix(iPtsMov)
    
    iPtsRef = addHomCoord2D(iPtsRef)
    iPtsMov = addHomCoord2D(iPtsMov)
    
    iPtsRef = iPtsRef.transpose()
    iPtsMov = iPtsMov.transpose()
    
    oMat2D = iPtsRef * iPtsMov.transpose() * np.linalg.inv(iPtsMov * iPtsMov.transpose())
    
    return oMat2D
````

3. Napišite funkcijo, ki poišče pare korespondenčnih točk med dvema množicama točk: 
```python
    def findCorrespondingPoints( iPtsRef, iPtsMov ):
        return iPtsRef_t, iPtsMov_t
```
 kjer je `iPtsRef` matrika s koordinatami referenčnih točk, `iPtsMov` pa matrika s koordinatami premičnih točk. Matriki imata dimenziji $K_R \times 2$ in $K_M \times 2$. Funkcija vrne sortirane koordinate `iPtsRef_t` in `iPtsMov_t` tako da istoležeče vrstice predstavljajo pare korespondenčnih točk.

````python
def findCorrespondingPoints(iPtsRef, iPtsMov):
    """Poisci korespondence kot najblizje tocke"""
    iPtsMov = np.array(iPtsMov)
    iPtsRef = np.array(iPtsRef)
    
    idxPair = -np.ones((iPtsRef.shape[0], 1), dtype='int32')
    idxDist = np.ones((iPtsRef.shape[0], iPtsMov.shape[0]))
    for i in range(iPtsRef.shape[0]):
        for j in range(iPtsMov.shape[0]):
            idxDist[i, j] = np.sum((iPtsRef[i, :2] - iPtsMov[j,:2])**2)
            
    while not np.all(idxDist == np.inf):
        i, j = np.where(idxDist == np.min(idxDist))
        idxPair[i[0]] = j[0]
        idxDist[i[0],:] = np.inf
        idxDist[:, j[0]] = np.inf
        
    idxValid, idxNotValid = np.where(idxPair >= 0)
    idxValid = np.array(idxValid)
    iPtsRef_t = iPtsRef[idxValid, :]
    iPtsMov_t = iPtsMov[idxPair[idxValid].flatten(), :]
    return iPtsRef_t, iPtsMov_t
````

4. Napišite funkcijo za poravnavo dveh oblik z metodo iterativno najbližje točke (ICP):
```python
    def alignICP2D( iPtsRef, iPtsMov, iEps, iMaxIter ):
        return oMat2D, oErr
```
kjer sta `iPtsRef` in `iPtsMov` matriki s koordinatami referenčnih in premičnih točk. Matriki imata dimenziji $K_R \times 2$ in $K_M \times 2$. `iEps` in `iMaxIter` določata pogoje za zaustavitev postopka, kjer je `iEps` najmanjša sprememba parametrov preslikave, `iMaxIter` pa maksimalno število korakov $k_{max}$. V funkciji uporabite 2D afino aproksimacijsko poravnavo, funkcija pa naj vrne optimalno 2D afino preslikavo med referenčnimi in premičnimi točkami v obliki homogene matrike `oMat2D` in vektor `oErr` velikosti $k_{max} \times 1$, ki podaja napako $\Sigma^2$ med korespondenčnimi točkami v vsakem koraku metode ICP. 

 Preverite pravilnost delovanja funkcije tako, da vse točke izbrane oblike preslikate s poljubnimi parametri 2D afine preslikave in preverite ali funkcija vrne preslikavo v `oMat2D`, ki je podobna dani vaši 2D afini preslikavi in da je končna napaka `oErr` manjša od začetne napake in približno enaka 0. Naložite referenčne in premične točke dveh poljubnih, a različnih oblik iz dane zbirke in preverite, da funkcija obliki ustrezno poravna.

````python
def alignICP(iPtsRef, iPtsMov, iEps=1e-6, iMaxIter=50, plotProgress=False):
    """Postopek iterativno najblizje tocke"""
    curMat = []; oErr = []; iCurIter = 0
    if plotProgress:
        iPtsMov0 = np.matrix(iPtsMov)
        fig = plt.figure()
        ax = fig.add_subplot(111)
    while True:
        iPtsRef_t, iPtsMov_t = findCorrespondingPoints(iPtsRef, iPtsMov)
        oMat2D = mapAffineApprox2D(iPtsRef_t, iPtsMov_t)
        iPtsMov = np.dot(addHomCoord2D(iPtsMov), oMat2D.transpose())
        curMat.append(oMat2D)
        oErr.append(np.sqrt(np.sum(iPtsRef_t[:,:2] - iPtsMov_t[:,:2])**2))
        iCurIter = iCurIter + 1
        dMat = np.abs(oMat2D - transAffine2D())
        if iCurIter > iMaxIter or np.all(dMat <iEps):
            break
    
    oMat2D = transAffine2D()
    for i in range(len(curMat)):
        
        if plotProgress:
            iPtsMov_t = np.dot(addHomCoord2D(iPtsMov0), oMat2D.transpose())
            ax.clear()
            ax.plot(iPtsRef[:,0], iPtsRef[:,1], 'ob')
            ax.plot(iPtsMov[:,0], iPtsMov[:,1], 'om')
            fig.canvas.draw()
            plt.pause(1)
        oMat2D = np.dot(curMat[i], oMat2D)
    return oMat2D, oErr
````
